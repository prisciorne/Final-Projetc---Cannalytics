```python
import requests
from bs4 import BeautifulSoup

url = 'https://wisevoter.com/country-rankings/gini-coefficient-by-country/'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
table = soup.find('table')
headers = [th.text.strip() for th in table.find_all('th')]

rows = []
for row in table.find_all('tr'):
    data = [td.text.strip() for td in row.find_all('td')]
    rows.append(data)

df = pd.DataFrame(rows[1:], columns=headers)
print(df)
```

```python
df.to_excel('gini.xlsx', index=False)
```

```python
url = "https://en.wikipedia.org/wiki/List_of_sovereign_states_by_tax_revenue_to_GDP_ratio"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
table = soup.find("table", class_="wikitable")

data = []
header_row = table.find("tr")
headers = [header.text.strip() for header in header_row.find_all("th")]
rows = table.find_all("tr")[1:]
for row in rows:
    row_data = [cell.text.strip() for cell in row.find_all("td")]
    data.append(dict(zip(headers, row_data)))

df4 = pd.DataFrame(data)

# Print the DataFrame
print(df4)
```

```python
df4.rename(columns={'Country': 'country'}, inplace=True) 
df4.rename(columns={'Taxrevenue(billions)': 'totalrev'}, inplace=True) 
```

```python
df4.head()
```

```python
data2['totalrev'] = data2['country'].map(df4.set_index('country')['totalrev'])
data2.head()
```
